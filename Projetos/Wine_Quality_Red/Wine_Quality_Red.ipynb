```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
wine = pd.read_csv('winequality-red.csv', sep=';')
```

```python
wine.head()
```

```python
wine.info()
```

```python
wine.corr()["quality"].sort_values(ascending = False)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
train_set, test_set = train_test_split(wine, test_size=0.2, random_state=42)
x_train = train_set.drop("quality", axis=1)
x_train_scaled = scaler.fit_transform(x_train)
y_train = train_set["quality"]
x_test = test_set.drop("quality", axis=1)
x_test_scaled = scaler.transform(x_test)
y_test = test_set["quality"]
```

```python
from sklearn.neural_network import MLPClassifier

mlpc = MLPClassifier(max_iter=2000, random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'hidden_layer_sizes': [
        (50,),               # 1 camada oculta com 50 neurônios
        (100,),              # 1 camada oculta com 100 neurônios
        (100, 50),           # 2 camadas: 100 + 50 neurônios
        (150, 100, 50),      # 3 camadas: 150 + 100 + 50 neurônios
    ],
    'activation': ['relu', 'tanh', 'logistic'],
    'solver': ['adam', 'sgd'],
    'alpha': [0.0001, 0.001],
    'learning_rate_init': [0.001, 0.01]
}

gridSearch = GridSearchCV(mlpc, param_grid, cv = 5, scoring="accuracy", n_jobs=-1)
```

```python
gridSearch.fit(x_train_scaled, y_train)
```

```python
gridSearch.best_score_
```

```python
mlpc_best = gridSearch.best_estimator_
y_predict = mlpc_best.predict(x_test_scaled)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_predict, y_test)
print(accuracy)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test, y_predict)

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=sorted(y_test.unique()))
disp.plot(cmap='Blues')
plt.title("Confusion Matrix")
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(y_test, y_predict)
print(f"Mean Absolute Error: {mae:.2f}") #Acurácia de 64% pode parecer baixa, mas utilizando o MAE podemos ver que o modelo
                                        #erra em média apenas 0.38 pontos da verdadeira nota do vinho
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, y_predict)
print(f"Mean Squared Error: {mse:.2f}")
```
